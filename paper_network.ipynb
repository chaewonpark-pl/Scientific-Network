```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

paper_df = pd.read_csv("Paper_Dataset_Bio.csv")

# Step 2: Filter the data for '과제수행년도' = 2020
paper_df_2019 = paper_df[paper_df['과제수행년도'] == 2019]

# Step 2: Create a Network from 'Affiliation_KR' column
affiliation_list = paper_df_2019['Affiliation_KR'].dropna()

# Create the graph
G = nx.Graph()

# Step 3: Add edges based on co-occurrence of organizations
for affiliation in affiliation_list:
    organizations = affiliation.split(';')
    for i in range(len(organizations)):
        for j in range(i+1, len(organizations)):
            org1 = organizations[i].strip()
            org2 = organizations[j].strip()
            G.add_edge(org1, org2)

# Step 4: Calculate Centrality
centrality = nx.degree_centrality(G)

# Get top 10 organizations by centrality
top_10_orgs = sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:10]
top_10_orgs

# Step 5: Adjust Node Size and Visualize Network (Without Labels)
plt.figure(figsize=(12, 12))

# Adjust node sizes (scaled down to 1/10)
node_sizes = [centrality[node] * 1000 for node in G.nodes]

# Draw the graph without labels
pos = nx.spring_layout(G, k=0.3, seed=42)
nx.draw(G, pos, with_labels=False, node_size=[size/10 for size in node_sizes], font_size=10, font_weight='bold', node_color='skyblue')

# Display the top 10 organizations
print("Top 10 Centrality Organizations:")
for org, score in top_10_orgs:
    print(f"{org}: {score}")

# Title and visualization
plt.title("Paper Network in 2019", fontsize=16)
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from itertools import combinations

paper_df = pd.read_csv("Paper_Dataset_Bio_2.csv")

# Step 1: Filter the data for '과제수행년도' = 2020
paper_df_2020 = paper_df[paper_df['과제수행년도'] == 2020]

# Step 2: Create a Network from 'Affiliation_KR' column
affiliation_list = paper_df_2020['Affiliation'].dropna()

# Create the graph
G = nx.Graph()

# Step 3: Add edges based on co-occurrence of organizations
for affiliation in affiliation_list:
    organizations = affiliation.split(';')
    for i in range(len(organizations)):
        for j in range(i+1, len(organizations)):
            org1 = organizations[i].strip()
            org2 = organizations[j].strip()
            G.add_edge(org1, org2)

# Step 4: Calculate Centrality
centrality = nx.degree_centrality(G)

# Get top 10 organizations by centrality
top_10_orgs = dict(sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:30])

# Step 5: Adjust Node Size and Visualize Network (With Labels for Top 10 Only)
plt.figure(figsize=(12, 12))

# Adjust node sizes (scaled down to 1/10)
node_sizes = [centrality[node] * 1000 for node in G.nodes]

# Position the nodes
pos = nx.spring_layout(G, k=0.3, seed=42)

# Draw nodes and edges
nx.draw_networkx_nodes(G, pos, node_size=[size/10 for size in node_sizes], node_color='skyblue')
nx.draw_networkx_edges(G, pos, alpha=0.3)

# Draw labels only for top 10 organizations
nx.draw_networkx_labels(G, pos, labels={node: node for node in top_10_orgs.keys()}, font_size=8, font_color="black")

# Display the top 10 organizations
print("Top 10 Centrality Organizations:")
for org, score in top_10_orgs.items():
    print(f"{org}: {score}")

# Title and visualization
plt.title("Paper Network in 2020 (Top Organizations Only)", fontsize=16)
plt.show()
```
